```python
import numpy as np
from pandas import read_csv    

import time

from datetime import datetime


#target_folder="./time-data/evaluation/"
target_folder="./time-data/setup/"


sensors=["fridge"]
#sensors=["microwave","fridge","cubiertos","ollas"]
tags=["1","2"] #,"3","4","5","6"]


#labels=["None","1","2","3","4","5","6"]
labels=["None","1","2"]

y_true=[]
y_pred=[]

for i,sensor in enumerate(sensors):
    print("METRIC OF SENSOR ",sensor)
    gt=read_csv(target_folder+"gt-labelling-sensors/"+sensor+".tsv", sep="\t",parse_dates=True, na_values=0)
    sv=read_csv(target_folder+sensor+".matrix.tsv", sep="\t",parse_dates=True,decimal=",")
    print(gt.dtypes)
    print(gt)
    print(gt["TIME"])
    print(sv)
    print(sv["TIME"])

    gt=gt.fillna(0)
    for tx,tt in enumerate(gt["TIME"]):
        
        if(sum(gt.values[tx][2:])==0):
            maxI=0
        else:
            maxI=np.argmax(gt.values[tx][2:])+1
        #print("GT:",tx,tt, "user:",gt.values[tx][2:],maxI)
        y_true.append(maxI)
        
        
        
        tt2=datetime.strptime(tt, '%Y-%m-%d %H:%M:%S').strftime('%Y-%m-%d %H:%M:%S')
        print(tt,tt2,np.where(sv["TIME"]==tt2))
        ty=np.where(sv["TIME"]==tt2)[0][0]
        if(sum(sv.values[ty][1:])==0):
            maxJ=0
        else:
            maxJ=np.argmax(sv.values[ty][1:])+1
        #print("PV:",ty,tt2, "user:",sv.values[ty][1:],maxJ)
        y_pred.append(maxJ)
        if(maxJ!=maxI):
            print(tt)
        
```

```python
print(y_true)
print(y_pred)

print(np.array(y_true).shape)
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

cm = confusion_matrix(y_true, y_pred)
print(cm)
print('Classification Report')

print(classification_report(y_true, y_pred, target_names=labels))
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
disp.plot(cmap=plt.cm.Blues)
plt.show()
```
